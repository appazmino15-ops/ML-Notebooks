```python
#Funcion

def neurona_MP_alarmas(inputs,pesos,threshold):
  suma=0

  for i in range(len(inputs)):
    suma += inputs[i]*pesos[i]

  if suma >= 1:
      return 1
  else:
      return 0
```

```python
# Escenario 1: humo y temperatura activados
print(neurona_MP_alarmas([1, 1, 0], [0.8, 0.5, 0.3], 0.7))

# Escenario 2: solo movimiento detectado
print(neurona_MP_alarmas([0, 0, 1], [0.8, 0.5, 0.3], 0.7))

# Escenario 3: humo apagado, temperatura alta, movimiento detectado
print(neurona_MP_alarmas([0, 1, 1], [0.8, 0.5, 0.3], 0.7))

# Escenario 4: humo apagado, temperatura alta, movimiento apagado
escenario4 = [0, 1, 0]
print(neurona_MP_alarmas([0, 1, 0], [0.8, 0.5, 0.3], 0.7))

# Escenario 5: Todo apagado
escenario5 = [0,0,0]
print(neurona_MP_alarmas(escenario5,[0.8,0.5,0.3],0.7))
```
